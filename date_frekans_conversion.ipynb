```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_excel("/Users/neseuygun/Desktop/data/exchange_rate.xlsx")
```

```python
df
```

```python
df.rename(columns={"Tarih":"Date", "Dolar":"Dollar"}, inplace=True)
```

```python
df.head()
```

```python
#change type of date as datetime
df["Date"] = df["Date"].apply(lambda x: datetime.strptime(x, "%d-%m-%Y"))
```

```python
print(df.head(),"\n")
print(df.dtypes)
```

```python
#convert date columns into index
df.set_index("Date", inplace=True)
df
```

```python
# Convert date into weekly data structure
df_new = round(df.resample("W").mean(),2)
df_new
```

```python
# Convert data into monthly data structure
df_new = round(df.resample("M").mean(),2)
df_new
```

```python
# Convert data into quarterly data structure
df_new = round(df.resample("Q").mean(),2)
df_new
```

```python
# Convert data into yearly data structure
df_new = round(df.resample("Y").mean(),2)
df_new
```
