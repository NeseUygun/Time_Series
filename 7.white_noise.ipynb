```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from arch.unitroot import ADF, PhillipsPerron, KPSS
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# For white noise, our data should has normal distribution that it mean equals to zero and variance equals to one.
data = np.random.randn(1000)
data
```

```python
plt.plot(np.arange(len(data)), data)
plt.title("White Noise Process")
plt.show()
```

```python
# If we split this data into different groups, we expect that each group has similar mean and variance.
samples = np.split(data,20)
mean_, variance_ = [], []
for i in samples:
    mean_.append(np.mean(i))
    variance_.append(np.var(i))

plt.plot(np.arange(len(mean_)), [data.mean()]*len(mean_), label="Mean", color="blue")
plt.scatter(x=np.arange(len(mean_)), y=mean_, color="blue")

plt.plot(np.arange(len(variance_)), [data.var()]*len(variance_), label="variance", color="red")
plt.scatter(x=np.arange(len(variance_)), y=variance_, color="red")
plt.legend()
plt.show()
```

As we can see in the above graph, we are able to observe samples means are disributed around zero and samples variances are distributed around one.

```python
#Now we can check if our series is stationary or not with autocorrelation graph. 
acf = plot_acf(data)
plt.show()
```
