```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = 0
step = 100
direction = ["Forward", "Backward"]
location = [y]

for i in range(1, step + 1):
    step = random.choice(direction)

    if step == "Forward":
        y+=1
    else:
        y-=1
    location.append(y)

print(location)
             
```

```python
plt.plot(location)
plt.show()
```

```python
def random_walk(n):
    x, y = 0, 0
    time_ = np.arange(n+1)
    direction = ["forward", "backward"]
    location = [y]
    

    for i in range(1, n+1):
        step = random.choice(direction)
        if step == "forward":
            y+=1
        else:
            y-=1
        location.append(y)
    print(time_)
    print(len(time_))
    print(location)
    print(len(location))
    return time_, location
    

x, y = random_walk(1000)
plt.plot(x, y)
plt.show()
```
