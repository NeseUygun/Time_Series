```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import yfinance as yf
import evds as e
```

```python
#1. Leptokurtic Distribution
stock = yf.download("XU100.IS", period="2Y", interval="1d")["Close"]
stock = pd.DataFrame(stock)
stock["Return"] = stock["Close"].pct_change()
stock
```

```python
# Let's plot the closing price and see if it makes any sense for us
sns.histplot(stock.Close, kde=True, bins=30)
plt.show()
```

```python
# Now plot return of the closing price and see if it makes any sense for us
sns.histplot(stock.Return, kde=True, bins=30)
plt.show()
```

As we can see at the return of the closing price plot, we have leptokurtic distribution which is distributions with positive kurtosis larger than that of a normal distribution. A normal distribution has a kurtosis of exactly three.
In general, leptokurtic distributions have heavier tails or a higher probability of extreme outlier values. It means there is high standart deviation so risk is high.

# 2. LogNormal Distribution
If a series has not a normal distribution but its logarithm has a normal distribution we are called it LogNormal Distribution.
What is difference between LogNormal Distribution and Normal Distribution?
The normal distribution is symmetrical, whereas the lognormal distribution is not. Because the values in a lognormal distribution are positive,

Let's try to explain why lognormal values is just positive with an example. whereas return of security could be negative but price of security couldn't.
So we can not take logarithm of return of security.

There is an assumption that if a security's logaritmic return has a normal distribution, its simple return has a lognormal distribution.

```python
# Let's create a series that has lognoral distribution and plot it
y = np.random.lognormal(3,1,1000) #series mean is 3, variance is 1
sns.histplot(y, kde=True)
plt.show()
```

```python
# Now let's take logarithm of y and see if it has a normal distribution by plotting
x = np.log(y)
sns.histplot(x, kde=True)
plt.show()
```

```python
# if we have x series that has normal distribution like above and want to get y by calculating the exponential value of x
y = np.exp(x)
sns.histplot(y, kde=True)
plt.show()
```
