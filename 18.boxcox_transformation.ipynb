```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import boxcox, yeojohnson
```

```python
#Box Cox
data = np.random.lognormal(0,1,1000)
bc, lambda_param = boxcox(data)
```

```python
fix, ax = plt.subplots(2,2)
ax[0,0].plot(data)
sns.histplot(data, kde=True, ax=ax[0,1])
ax[1,0].plot(bc)
sns.histplot(bc, kde=True, ax=ax[1,1])
plt.show()
```

```python
# Yeo Johnson
data = np.random.lognormal(0,1,1000)*-1
yj, lambda_param = yeojohnson(data)
```

```python
fix, ax = plt.subplots(2,2)
ax[0,0].plot(data)
sns.histplot(data, kde=True, ax=ax[0,1])
ax[1,0].plot(yj)
sns.histplot(yj, kde=True, ax=ax[1,1])
plt.show()
```
